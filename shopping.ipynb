# 11번가 main banner crawling

```python
#-*- coding: utf-8 -*-

import requests
import urllib3
import json
from urllib.parse import urlencode
from bs4 import BeautifulSoup

from PIL import Image
import requests
import matplotlib.pyplot as plt

urllib3.disable_warnings()
```

```python
## 11번가 메인 베너

base_url = 'http://m.11st.co.kr/MW/html/main.html#/home'
req = requests.get(base_url, verify=False)
soup = BeautifulSoup(req.text, "html.parser")
scripts = soup.findAll('script', {'type':'text/javascript'})        

jsonObj = None
no = 0;

for s in scripts:
    if 'var homeBillBannerList' in s.text:
        script = s.text
        script = script.split('var homeBillBannerList = ')[1]
        script = script.split(';')[0]
        #script = script.split('var homeBillBnrEl ')[0]
        #print(script)
        
        jsonStr = script
        jsonList = json.loads(jsonStr)
        
        for banner in jsonList:
            no = no + 1;
            url = str(banner['homeBillBanner']['lnkBnnrImgUrl'])
            print('[', no, '] ', url)
            
            response = requests.get(url, stream=True, verify=False)
            img = Image.open(response.raw)
            plt.imshow(img)
            plt.show()
```
